```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
record_df = pd.read_csv("record.csv")
record_df.head()
```

```python
r_epochs = len(record_df['epoch'])
print(r_epochs)
```

```python
plt.title("Discriminator loss ")
plt.plot(np.arange(0, r_epochs), record_df['d_Loss'], 'r') # c
plt.legend(['d_loss'], loc='lower right')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.grid()
#plt.savefig("Class_Acc.jpg", dpi=100)
plt.show()
```

```python
plt.title("Discriminator Acc ")
plt.plot(np.arange(0, r_epochs), record_df['accuracy'], 'y') # c
plt.legend(['accuracy'], loc='lower right')
plt.ylabel('Acc')
plt.xlabel('Epochs')
plt.grid()
#plt.savefig("Class_Acc.jpg", dpi=100)
plt.show()
```

```python
plt.title("Generator loss")
plt.plot(np.arange(0, r_epochs), record_df['g_loss'], 'y') # c
plt.legend(['g_loss'], loc='lower right')
plt.ylabel('g loss')
plt.xlabel('Epochs')
plt.grid()
#plt.savefig("Class_Acc.jpg", dpi=100)
plt.show()
```

```python


plt.title('adv, recon, id Losses')
#plt.plot(np.arange(0, r_epochs), record_df['adv'], 'g')
plt.plot(np.arange(0, r_epochs), record_df['recon'], 'y') # c
plt.plot(np.arange(0, r_epochs), record_df['id'], 'c') # c

plt.legend(['adv', 'recon', 'id'], loc='upper right')
plt.ylabel('Adv, recon, id losses')
plt.xlabel('Epochs')
plt.grid()
#plt.savefig("Class_loss.jpg", dpi=100)
plt.show()

```
